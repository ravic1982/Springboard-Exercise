```python
def range_of_hr(var1):
    if (var1 >= 4) & (var1 < 12):
        a = 'Morning'
    elif (var1 >= 12) & (var1 < 17):
        a = 'Afternoon'
    elif (var1 >= 17) & (var1 < 22):
        a = 'Evening'
    else:
        a = 'Night'
    return a  
```

```python
def dow(val):
    if (val == 0):
        d = 'Sunday'
    elif (val == 1):
        d = 'Monday'
    elif (val == 2):
        d = 'Tuesday'    
    elif (val == 3):
        d = 'Wednesday'
    elif (val == 4):
        d = 'Thursday' 
    elif (val == 5):
        d = 'Friday'
    elif (val == 6):
        d = 'Saturday'  
    return d       
```

```python
orders_df['range_of_hr'] = orders_df.order_hour_of_day.apply(range_of_hr)
orders_df.head()
```

```python
orders_df['weekday'] = orders_df.order_dow.apply(dow)
orders_df.head()
```
